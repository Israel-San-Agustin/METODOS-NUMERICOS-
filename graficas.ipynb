# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return (x+1)**(1/2)
def g(x):
    return 1+ x*(1/2)
def h(x):
    return 1+ x*(1/2)-(x**2)*(1/8)
def d(x):
    return 1+ x*(1/2)-(x**2)*(1/8)+(x**3)*(1/16)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10, 10,100)
```

```python
f(-5),f(1)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(5,5)) #tama침o de la figura
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, d(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.xlim(-10, 10)
plt.ylim(-5, 5)
plt.title("suma de funciones")
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.grid()
plt.show()
```

se muestra como pudes graficar diferentes funciones en un mismo plano ademas de que los codigos nos permiten cambiar de tama침o asi como tambien el como agregar titulos y darnos una manera de graficar en python
