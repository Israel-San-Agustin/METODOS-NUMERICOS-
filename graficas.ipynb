# Ejemplo: Polinomio de Maclaurinn de grado tres con  $f(x)= e^{x/2}sen(x/3)$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return np.exp(x/2)*np.sin(x/3)
def g(x):
    return (1/3)*x
def h(x):
    return (1/3)*x+(1/6)*x**(2)
def d(x):
    return (1/3)*x+(1/6)*x**(2)+(23/648)*x**(3)
```

```python
#valores del eje x que toma la grafica
x=np.linspace(-10, 10, 100)
```

```python
# Graficar ambas funciones.
#plt.figure(figsize=(5,5)) #tamaño de la figura
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, d(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.xlim(-10, 10)
plt.ylim(-5, 5)
plt.title("suma de funciones")
plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid()
plt.show()
```
