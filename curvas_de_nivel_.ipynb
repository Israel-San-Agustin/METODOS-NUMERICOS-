```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funcin f(x, y) = 1 + x
def f(x, y):
    return 1 + x

# Crear malla de valores para x e y
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Crear figura en 3D
fig = plt.figure(figsize=(12, 6))

# Subplot 3D para la superficie y las intersecciones
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Interseccin con el plano x = 0 (lnea horizontal en z = 1)
y_line = np.linspace(-5, 5, 10)
z_line = np.ones_like(y_line)
ax.plot([0] * len(y_line), y_line, z_line, 'r-', linewidth=2, label="Interseccin x=0")

# Interseccin con el plano y = 0 (recta en el plano xz)
x_line = np.linspace(-5, 5, 10)
z_line = f(x_line, 0)
ax.plot(x_line, [0] * len(x_line), z_line, 'b-', linewidth=2, label="Interseccin y=0")

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Superficie y sus intersecciones")
ax.legend()

# Subplot de curvas de nivel
ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, Z, levels=10, cmap="viridis")
ax2.clabel(contour, inline=True, fontsize=8)
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_title("Mapa de curvas de nivel")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funcin f(x, y)
def f(x, y):
    return np.sqrt(np.maximum(4 - 4*x**2 - y**2, 0))  # Asegurar valores reales

# Crear una malla de valores para x e y
x = np.linspace(-1, 1, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Crear la figura y los ejes
fig = plt.figure(figsize=(12, 6))

# --- Mapa de curvas de nivel ---
ax1 = fig.add_subplot(1, 2, 1)
contour = ax1.contour(X, Y, Z, levels=10, cmap="viridis")
ax1.clabel(contour, inline=True, fontsize=8)
ax1.set_title("Mapa de curvas de nivel")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.grid(True)

# --- Grfica 3D con intersecciones ---
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z, cmap="viridis", alpha=0.8)

# Interseccin con planos z = k
k_values = [0.5, 1, 1.5]  # Diferentes valores de z para interseccin
colors = ['r', 'g', 'b']

for k, color in zip(k_values, colors):
    ax2.contour3D(X, Y, Z, levels=[k], colors=color)

ax2.set_title("Grfica 3D con intersecciones de planos")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("z")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funcin f(x,y) = |x| + |y|
def f(x, y):
    return np.abs(x) + np.abs(y)

# Crear la malla de valores
gx = np.linspace(-2, 2, 100)
gy = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(gx, gy)
Z = f(X, Y)

# Crear la figura y los ejes 3D
fig = plt.figure(figsize=(12, 6))

# Subplot 1: Mapa de curvas de nivel
ax1 = fig.add_subplot(1, 2, 1)
contour = ax1.contourf(X, Y, Z, levels=20, cmap='plasma')
plt.colorbar(contour, ax=ax1, label='Altura Z')
ax1.set_title('Mapa de curvas de nivel')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# Subplot 2: Grfica 3D con intersecciones de los planos
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='plasma', alpha=0.8, edgecolor='k')
ax2.contour3D(X, Y, Z, levels=20, cmap='coolwarm')
ax2.view_init(elev=20, azim=60)  # Nueva perspectiva para mejor visualizacin
ax2.set_title('Grfica 3D con intersecciones de planos')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

plt.tight_layout()
plt.show()
```
