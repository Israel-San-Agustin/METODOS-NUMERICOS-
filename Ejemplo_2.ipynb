# Grafica de $f(x) = (x-1)^{1/2}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#funcion cuadratica.
def f(x):
    return (x+1)**(1/2)
def g(x):
    return 1+x*(1/2)
def h(x):
    return 1+x*(1/2)-(x**2)*(1/8)
def d(x):
    return 1+x*(1/2)-(x**2)*(1/8)+(x**3)*(1/16)
```

```python
#valores del eje x que toma el grafico
x=np.linspace(-10,10,100)
```

```python
#grafica de las funciones
plt.figure(figsize=(5,5)) #tamaño de la figura
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, d(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title("$f(x) = (x-1)^{1/2}$")
plt.savefig("output.png") # Guardar gráfico como imágen PNG.
plt.grid()
plt.show()
```
